## Dataset Overview

This dataset contains aviation accident and incident records.  
It includes information such as event date, location, aircraft type,
injury severity, and probable cause.

The dataset will be used to explore trends in aviation accidents and
identify factors associated with higher injury severity.

```python
key_cols = [
    "Event.Date",
    "Injury.Severity",
    "Country",
    "Aircraft.Category",
    "Make",
    "Model"
]

df[key_cols].isna().sum()
```

```python
df["Event.Date"] = pd.to_datetime(df["Event.Date"], errors="coerce")
df["Year"] = df["Event.Date"].dt.year
```

```python
df["Event.Date"].dtype
```

```python
df_clean = df.dropna(subset=["Event.Date", "Injury.Severity", "Country", "Year"])
```

```python
df_clean.shape
```

## Data Cleaning & Preparation

The dataset was cleaned by identifying missing values and removing records
with missing critical information such as event date, injury severity, and
country.

The Event.Date column was converted to a datetime format, and additional
time-based features (year and month) were created to support trend analysis.

```python
accidents_by_year = (
    df_clean["Year"]
    .value_counts()
    .sort_index()
)
```

```python
injury_distribution = df_clean["Injury.Severity"].value_counts()
```

```python
accidents_by_country = (
    df_clean["Country"]
    .value_counts()
    .head(10)
)
```

```python
df_clean = df_clean[
    (df_clean["Year"] >= 1980) &
    (df_clean["Year"] <= 2023)
]
```

```python
accidents_by_year = (
    df_clean["Year"]
    .value_counts()
    .sort_index()
)

accidents_by_year.plot(
    kind="line",
    title="Aviation Accidents Over Time",
    xlabel="Year",
    ylabel="Number of Accidents"
)
```

### Aviation Accidents Over Time

The visualization shows how aviation accident occurrences have changed over
the years. This trend can help aviation authorities assess whether safety
measures have improved or if additional interventions are needed.

```python
injury_distribution = df_clean["Injury.Severity"].value_counts()

injury_distribution.plot(
    kind="bar",
    title="Distribution of Injury Severity in Aviation Accidents",
    xlabel="Injury Severity",
    ylabel="Number of Accidents"
)
```

### Injury Severity Distribution

Most aviation accidents fall into specific injury severity categories. Understanding
this distribution helps safety regulators prioritize prevention strategies that
reduce severe and fatal outcomes.

```python
accidents_by_country = (
    df_clean["Country"]
    .value_counts()
    .head(10)
)

accidents_by_country.plot(
    kind="bar",
    title="Top 10 Countries with Highest Aviation Accident Counts",
    xlabel="Country",
    ylabel="Number of Accidents"
)
```

### Countries with Highest Accident Counts

The chart highlights the countries with the highest number of recorded aviation
accidents. This insight can guide regional safety audits and targeted training
programs.

## Recommendations

1. Aviation authorities should continue monitoring long-term accident trends to
evaluate the effectiveness of existing safety regulations.

2. Safety programs should focus on reducing high-severity injuries by addressing
common causes identified in serious accident categories.

3. Countries with higher accident counts may benefit from targeted safety
interventions, improved reporting systems, and additional pilot training.
